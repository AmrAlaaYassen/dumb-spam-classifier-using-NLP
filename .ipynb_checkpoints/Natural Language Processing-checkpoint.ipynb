```python
def text_process(mess):
    
    non_punc = [char for char in mess if char not in string.punctuation]
    
    non_punc = ''.join(non_punc)
    
    return [word for word in non_punc.split() if word.lower() not in stopwords.words('english')]
```

```python
string.punctuation
```

```python
messages['message'].head(5)
```

```python
bag_of_words_transformer = CountVectorizer(analyzer=text_process).fit(messages['message']) # msh fahm aywe 
```

```python
len(bag_of_words_transformer.vocabulary_)
```

```python
mes = messages['message'][3]
```

```python
mes = bag_of_words_transformer.transform([mes])
```

```python
print(mes)
```

```python
messages_bow = bag_of_words_transformer.transform(messages['message'])
```

```python
messages_bow.nnz #check non zeros
```

```python
tf_idf_transformer = TfidfTransformer().fit(messages_bow)
```

```python
print(tf_idf_transformer)
```

```python
messages_tf_idf = tf_idf_transformer.transform(messages_bow)
```

```python
spam_detect_model = MultinomialNB().fit(messages_tf_idf,messages['label'])
```

```python
tfidf4 = tf_idf_transformer.transform(mes)
print(tfidf4)
```

```python
pipline = Pipeline([
    ('bow',CountVectorizer(analyzer=text_process)),
    ('tfidf',TfidfTransformer()),
    ('classifier',MultinomialNB())
])

msg_train,msg_test,label_train,label_test = train_test_split(messages['message'],messages['label'],test_size=0.3)

pipline.fit(msg_train,label_train)

pred = pipline.predict(msg_test)
print(classification_report(label_test,pred))
```
